```python
## Se importan los paquetes necesarios para trabajar.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set_style("whitegrid"); sns.set_context("paper")
import matplotlib as mpl
import os
```

```python
## Se carga el dataset 'equiposnac' creado a partir del archivo 'equiposnac2.csv'
equiposnac = pd.read_csv('equiposnac2.csv')
equiposnac = equiposnac.drop(['Unnamed: 0'],axis=1)
```

```python
## Se crea el campo 'pais_club2' que toma el valor True en caso que equiponac sea igual a pais_club (es decir en caso que
## el jugador milite en un club de su propio pais), False de lo contrario.
equiposnac['pais_club2'] = np.where(equiposnac['equiponac']==equiposnac['pais_club'],True,False)
```

```python
## Se construye la variable IJE (Indice de Jugadores en el Extranjero) que presenta en índice de jugadores que jugaban en el
## extranjero en un año dado, del total de jugadores convocados en una seleccion nacional, para el conjunto de paises de la
## CONMEBOL de manera agregada.

equiposnac_IJE = equiposnac.groupby('año').mean()
equiposnac_IJE['IJE'] = 1 - equiposnac_IJE['pais_club2']
equiposnac_IJE = equiposnac_IJE.drop('pais_club2',axis=1)
```

```python
## IJE del conjunto agregado de paises de la CONMEBOL
mpl.rc('lines', linewidth=4)
fig1, ax = plt.subplots( nrows=1, ncols=1 )
ax.plot(equiposnac_IJE,color='cyan');
plt.xlabel('Años');
plt.ylabel('IJE');
plt.gca().legend(['CONMEBOL']);
plt.show();
plt.close(fig1)
```
