```python
## Se escrpean los datos del sitio web 'https://www.footballdatabase.eu/'. Se obtienen los registros de todos los jugadores
## convocados a la selección nacional de su país en la CONMEBOL entre 1994 y 2019. Los campos que se obtienen son: año,
## equiponac (equipo nacional al que pertenece), jugador (nombre del jugador) y club en el que jugaba al concluir el año.

with requests.Session() as s:
    url = 'https://www.footballdatabase.eu/'
    r = s.get(url)
    soup = BeautifulSoup(r.text,'html.parser')
    r = s.post(url,data=login_data)
    clubes_raw = []
    jugadores_raw = []
    anio = []
    pais = []
    equipos = ['Bolivia','Colombia','Argentina','Ecuador','Uruguay','Peru','Chile','Venezuela','Paraguay','Brasil']
    cont_equipo = -1
        
    for pagina in paginas:
        anios = 1993
        cont_equipo += 1
        
        for ele in pagina:
            url = '{}{}'.format(url_base,ele)
            r = s.get(url)
            soup = BeautifulSoup(r.text,'html.parser')
            anios += 1
            sleep(1)
            
            for listclubs_class in soup.find_all(class_='listclubs'): 
                clubes_raw.append((listclubs_class.text, listclubs_class.next_sibling))

            for long_class in soup.find_all(class_='long'):
                jugadores_raw.append((long_class.text, long_class.next_sibling))
                for i in range(len(long_class)):
                    anio.append(anios)
                    
                for i in range(len(long_class)):
                    pais.append(equipos[cont_equipo])
                    
    r.close()
                               
jugadores = []
for i in range(len(jugadores_raw)):
    jugadores.append(jugadores_raw[i][0])

clubes_raw2 = []
clubes_raw = clubes_raw[1:]
for i in range(len(clubes_raw)):
    clubes_raw2.append(clubes_raw[i][0])

clubes_raw4 = []
pat = r'(?:Castilla|Sevilla Atlético|Valence B| (.+?)\xa0)'
for i in range(len(clubes_raw2)):
    clubes_raw3 = re.findall(pat,clubes_raw2[i])
    clubes_raw4.append(clubes_raw3)

clubes = []
for sublist in clubes_raw4:
    for i in sublist:
        clubes.append(i)       

equipos = pd.DataFrame({'año':anio,'equiponac':pais,'jugador':jugadores,'club':clubes})
equipos['club'] = equipos['club'].apply(lambda x: x.strip())
```
